```python
def read_csv(file_name):
    f=open(file_name).read()
    string_list=f.split("\n")[1:]
    final_list=[]
    for row in string_list:
        int_fields=[]
        string_fields=row.split(",")
        for field in string_fields:
            int_fields.append(int(field))
        final_list.append(int_fields)
    return final_list
done_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
done_list[0:1]
```

```python
births_per_month={}
def month_births(list_of_lists):
    for a_list in list_of_lists:
        month =a_list[1]
        births=a_list[4]
        if month in births_per_month:
            births_per_month[month]=births_per_month[month]+births
        else:
            births_per_month[month]=births
month_births(done_list)
births_per_month
```

```python
cdc_day_births={}
def dow_births(list_of_lists):
    for a_list in list_of_lists:
        day_of_week =a_list[3]
        births=a_list[4]
        if day_of_week in cdc_day_births:
            cdc_day_births[day_of_week]=cdc_day_births[day_of_week]+births
        else:
            cdc_day_births[day_of_week]=births
dow_births(done_list)
cdc_day_births
```

```python
def calc_counts(data,column):
    common_dict={}
    for row in data:
        element_1=row[column]
        element_2=row[4]
        if element_1 in common_dict:
            common_dict[element_1]=common_dict[element_1]+element_2
        else:
            common_dict[element_1]=element_2
    return common_dict
cdc_year_births=calc_counts(done_list,0)
cdc_year_births
```

```python
cdc_month_births=calc_counts(done_list,1)
cdc_month_births
```

```python
cdc_dom_births=calc_counts(done_list,2)
cdc_dom_births
```

```python
cdc_dow_births=calc_counts(done_list,3)
cdc_dow_births
```

```python
maximum = max(cdc_dow_births, key=cdc_dow_births.get) 
print(maximum, cdc_dow_births[maximum])
```
