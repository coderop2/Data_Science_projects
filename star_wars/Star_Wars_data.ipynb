```python
import pandas as pd
import numpy as np
star_wars_data=pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars_data.head(10)
```

```python
star_wars_data.columns
```

```python
star_wars_data = star_wars_data[pd.notnull(star_wars_data["RespondentID"])]
star_wars_data.head()
```

```python
mapping={"Yes": True, "No": False}
star_wars_data["Have you seen any of the 6 films in the Star Wars franchise?"]=star_wars_data["Have you seen any of the 6 films in the Star Wars franchise?"].map(mapping)
star_wars_data["Do you consider yourself to be a fan of the Star Wars film franchise?"]=star_wars_data["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(mapping)
star_wars_data.head()
```

```python
star_wars_data["Which of the following Star Wars films have you seen? Please select all that apply."].value_counts()
```

```python
star_wars_data.shape
```

```python
movie_names=["Star Wars: Episode I  The Phantom Menace","Star Wars: Episode II Attack of the Clones","Star Wars: Episode III Revenge of the Sith","Star Wars: Episode IV A New Hope","Star Wars: Episode V The Empire Strikes Back","Star Wars: Episode VI Return of the Jedi"]
star_wars_data = star_wars_data.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"})
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars_data.columns[3:9]:
    star_wars_data[col] = star_wars_data[col].map(movie_mapping)
```

```python
star_wars_data.head(2)
```

```python
star_wars_data = star_wars_data.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })
star_wars_data[star_wars_data.columns[9:15]] = star_wars_data[star_wars_data.columns[9:15]].astype(float)
star_wars_data[star_wars_data.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(range(6), star_wars_data[star_wars_data.columns[9:15]].mean(),color="lightgreen")
```

```python
plt.bar(range(6), star_wars_data[star_wars_data.columns[3:9]].mean(),color="lightblue")
```

# Findings:-

View counts:-
It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.

Rankings:-
So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. As I suspected, it looks like the "original" movies are rated much more highly than the newer ones.

```python
males = star_wars_data[star_wars_data["Gender"] == "Male"]
females = star_wars_data[star_wars_data["Gender"] == "Female"]
plt.bar(range(6), males[males.columns[9:15]].sum(),color="green")
plt.show()

plt.bar(range(6), females[females.columns[9:15]].sum(),color="lightgreen")
plt.show()
```

```python
plt.bar(range(6), males[males.columns[3:9]].sum(),color="lightblue")
plt.show()

plt.bar(range(6), females[females.columns[3:9]].sum())
plt.show()
```

# Final Findings:-

Male/Female differences in favorite Star Wars movie and most seen movie
Interestingly, more males watches episodes 1-3, but males liked them far less than females did.
