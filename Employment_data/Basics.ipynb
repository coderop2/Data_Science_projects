```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
recent_grads=pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads=recent_grads.dropna()
```

```python
recent_grads.describe()
```

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
```

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size',y='Median',title='Sample_size VS Median', kind='scatter')
```

```python
recent_grads.plot(x='Men',y='Median',title='Men VS Median', kind='scatter')
```

```python
recent_grads.plot(x='Women',y='Median',title='Women VS Median', kind='scatter')
```

```python
recent_grads.plot(x='ShareWomen',y='Median',title='ShareWomen VS Median', kind='scatter')
```

# Findings:-

It can be infered from the scater graph of ShareWomen VS Median that the share of women who work full time and earn high is less as compared to the large share of women who work for full time but earn less.
A seen from the Women VS Median and Men VS Median that the large number of men and women who graduate earn less than 40000 as seen from the strong correlation.

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(0,4):
    ax = fig.add_subplot(2,2,r+1)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=60)
    ax.set_xlabel(cols[r])
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(2,2,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=60)
    ax.set_xlabel(cols[r])
```
